![Alternative_Text](Iris.png)

The Iris Dataset was used in R.A. Fisher's Classic 1936 paper. The use of multiple measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning (ML) Repository.

It includes 3 iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:
 - **ID**
 - **SepalLengthCm**
 - **SepalWidthCm**
 - **PetalLengthCm**
 - **PetalWidthCm**
 - **Species**

**IMPORTING IMPORTANT LIBRARIES (DEPENDENCIES) AND LOADING THE DATASET**

```python
import pandas as pd # Library for Data Analysis & Data Manipulation
import numpy as np # Library for Scientific Computing
import seaborn as sns # Library for Statistical Data Visualization
import matplotlib.pyplot as plt # Plotting Library for High-Quality Data Visualization & Exploratory Data Analysis (EDA)
import warnings # Library to control how warning messages are displayed

warnings.filterwarnings('ignore') # Suppress all warnings
```

```python
df = pd.read_csv("Iris_Dataset.csv")
```

**DATA CLEANING**

```python
# Drop 'Id' column from the dataframe as it is not adding any significance in our result
df.drop(columns='Id', inplace=True) 
```

```python
# Rename Columns
df.rename({ 'SepalLengthCm' : 'SepalLength', 'SepalWidthCm' : 'SepalWidth', 'PetalLengthCm' : 'PetalLength', 'PetalWidthCm' : 'PetalWidth'} ,axis=1, inplace=True)
```

```python
df.isnull().sum() # Checks for missing values in the dataframe
```

As, there are no missing or null values in the dataframe. It means Data is already cleaned.

```python
df # All the changes are saved in the dataframe
```

**DATA VISUALIZATION**

**2D-Pair Plots**

*Let's make 2D-Pair Plots to distinguish the differences between Sepal & Petal Length of Irides**

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue='Species', size=3)
plt.show()
```

**Observations**
 - All type of flowers are well-separable for PetalLength and PetalWidth
 - All type of irides are a bit well-separable for PetalWidth and SepalWidth 

**Heatmap**

*Let's make a heatmap which shows the correlation between different features*

```python
fig = plt.figure(figsize=(9,6))
sns.heatmap(df.drop("Species", axis=1).corr(), cmap='Blues', annot=True)
```

**Observations**
 - PetalLength and PetalWidth : Corr 0.96 : Having Highest Correlation
 - PetalLength and SepalLength : Corr 0.87 : Having 2nd Highest Correlation
 - PetalWidth and SepalLength : Corr 0.82 : Having fair enough Correlation

**Distribution Plots**

*Let's make distribution plots for each of these (numerical) features which are useful for visualizing the data-points within each feature and help us to understand spread, skewness and presence of any outliers*

```python
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
fig = plt.figure(figsize=(20,4))
i = 1

for col in columns:
    plt.subplot(1,4,i)
    sns.distplot(df[col])
    i = i+1

plt.show()
```

**Observations**
 - **SepalLength :** Maximum distribution is around 5 to 6
 - **SepalWidth :** Maximum distribution is around 3
 - **PetalLength :** Maximum distribution is around 4.5
 - **PetalWidth :** Maximum distribution is around 1.5

**Univariate Analysis of Features**

*Let's create distribution plots for each feature separately with separate curves for each species of the iris*

```python
sns.FacetGrid(df, hue='Species', height=4).map(sns.distplot, 'SepalLength').add_legend()
```

 - We cannot separate irides just using SepalLength

```python
sns.FacetGrid(df, hue='Species', height=4).map(sns.distplot, 'SepalWidth').add_legend()
```

 - We cannot separate irides just using SepalWidth

```python
sns.FacetGrid(df, hue='Species', height=4).map(sns.distplot, 'PetalLength').add_legend()
```

 - We can separate Iris-Setosa flowers just using PetalLength easily
 - Also, Iris-Versicolor & Iris-Virginica are partially separated

```python
sns.FacetGrid(df, hue='Species', height=4).map(sns.distplot, 'PetalWidth').add_legend()
```

 - We can see Iris-Setosa is well-separable using PetalWidth
 - Iris-Versicolor & Iris-Virginica are partially separable using PetalWidth

**Box-Plots**

*Let's create box-plots for visualizing the comparison of distribution of each feature across species*

```python
fig, axis = plt.subplots(2, 2, figsize=(20,10))

sns.boxplot(x='Species', y='PetalLength', data=df, ax=axis[0,0])
sns.boxplot(x='Species', y='PetalWidth', data=df, ax=axis[0,1])
sns.boxplot(x='Species', y='SepalLength', data=df, ax=axis[1,0])
sns.boxplot(x='Species', y='SepalWidth', data=df, ax=axis[1,1])

plt.show()
```

**Observations**
 - **Iris-Setosa:** It's usually having small features except SepalWidth
 - **Iris-Versicolor:** It's having average features
 - **Iris-Verginica:** It's having bigger features except SepalWidth

**Violin-Plots**

*Let's generate violin-plots to visualize the distribution of data and its density*

```python
fig, axis = plt.subplots(2, 2, figsize=(15,5))

sns.violinplot(x='Species', y='PetalLength', data=df, ax=axis[0,0])
sns.violinplot(x='Species', y='PetalWidth', data=df, ax=axis[0,1])
sns.violinplot(x='Species', y='SepalLength', data=df, ax=axis[1,0])
sns.violinplot(x='Species', y='SepalWidth', data=df, ax=axis[1,1])

plt.show()
```

**Observation**
 - Wider-Sections represent more Data-Points

**Conclusion of Exploratory Data Analysis (EDA) on Iris Dataset:**
 - Data is balanced.
 - Strong Correlation between **PetalLength** and **PetalWidth**
 - Iris-Setosa is easily separable from other irides, even using a single feature
 - Iris-Versicolor and Iris-Virginica are difficult to separate, just using a single feature
